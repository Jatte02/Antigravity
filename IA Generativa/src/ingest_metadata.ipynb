```python
import os
from dotenv import load_dotenv
from sqlalchemy import create_engine
```

```python
# Cargar las variables del archivo .env al entorno que estamos

load_dotenv()

# Extraemos los valores usando el nombre de la variable
aws_user = os.getenv('AWS_RDS_USER')
aws_pass = os.getenv('AWS_RDS_PASS')
aws_port = os.getenv('AWS_RDS_PORT')
aws_dbname = os.getenv('AWS_RDS_DBNAME')
aws_host = os.getenv('AWS_RDS_HOST')
```

```python
# Construir el connection string
database_url = f'postgresql://{aws_user}:{aws_pass}@{aws_host}:{aws_port}/{aws_dbname}'

# Crear el engine

engine = create_engine(database_url)
```
